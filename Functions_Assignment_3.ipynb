1. What is the difference between a function and a method in Python?

Answer: A**function** in Python is a block of reusable code that performs a specific task. It is defined using the **def** keyword and can be called independently.

A **method** is a function associated with an object. It is called on the **object** and can access or modify its data. Methods are defined inside classes.

**Key Difference**:

**Function**: Independent, not tied to an object.

**Method**: Dependent on an object and called using object.method().

Example:

# Function
def greet(name):  
    print(f"Hello, {name}!")  

greet("Uday")  # Calling the function

# Method
class Person:  
    def __init__(self, name):  
        self.name = name  

    def greet(self):  
        print(f"Hello, {self.name}!")  

p = Person("Bhaskar")  
p.greet()  # Calling the method  

* Here, greet( ) is a function, while p.greet( ) is a method tied to the Person object.

5. What are iterators in Python and how do they differ from iterables?

Answer: In Python, iterators and iterables are used for looping through data.

* **Iterable**: An object that contains data and can return an iterator using the iter() function. Examples: lists, tuples, sets, and dictionaries.

* **Iterator**: An object with the methods __iter__() and __next__() that allows traversing through all the elements one by one. Iterators are used to fetch values one at a time.

**Difference**:

* **Iterable**: Can be looped over but does not track iteration state.

* **Iterator**: Tracks iteration state and returns the next value using next().

**Example**:

# Iterable (list)

nums = [1, 2, 3]  

it = iter(nums)  # Convert to iterator  

print(next(it))  # Output: 1  

print(next(it))  # Output: 2  



6. Explain the concept of generators in Python and how they are defined.

Answer: In Python, generators are special functions that yield values one at a time using the yield keyword instead of return. They are used to create iterators in a memory-efficient way.

**Features**:

1. **Lazy evaluation**: Values are generated on-the-fly, saving memory.

2. **State retention**: Generators retain their state between calls.

3. **Efficient looping**: Used for handling large data efficiently.

**Example: define a generator**

def count_up(n):  
    for i in range(n):  
        yield i     # Yields one value at a time  

gen = count_up(3)  

print(next(gen))  # Output: 0  

print(next(gen))  # Output: 1  

* Here, "yield" makes "count_up()" a generator function.

7. What are the advantages of using generators over regular functions?

Answer: **Advantages of Using Generators Over Regular Functions:**  

1. **Memory Efficiency:**  
   - Generators use lazy evaluation, producing values one at a time.  
   - Regular functions return all values at once, consuming more memory.  

2. **Faster Execution:**  
   - Generators yield values immediately, speeding up processing.  
   - Regular functions create and return entire lists, which takes more time.  

3. **Infinite Sequences:**  
   - Generators can produce infinite data streams without consuming all memory.  
   
4. **Simplified Code:**  
   - Generators make code cleaner and more readable by using "yield".  

5. **State Retention:**  
   - Generators retain their state, making them ideal for iterative tasks.

   **Example:**

   
   * Generator to yield squares of numbers

def squares(n):  
    for i in range(n):  
        yield i * i  

* Regular function returning a list

def squares_list(n):  
    return [i * i for i in range(n)]  


print(list(squares(5)))       # Memory-efficient  

print(squares_list(5))         # Stores entire list

9. 9. Explain the purpose and usage of the `map()` function in Python.

Answer: **'map()' Function in Python:**  

The 'map()' function applies a given function to **each item** of an iterable (e.g., list, tuple) and returns an **iterator** with the results.  

**Syntax:**  

  "map(function, iterable)"  


**Purpose and Usage:**  

1. **Applies a function** to all elements of an iterable without using explicit loops.  

2. **Memory-efficient** because it returns an iterator instead of a list.  

3. Often used with **lambda functions** for concise operations.  

**Example:**  

* Using map() to double numbers
nums = [1, 2, 3, 4]  

doubled = map(lambda x: x * 2, nums)  

print(list(doubled))  # Output: [2, 4, 6, 8]  

* Here, 'map()' applies the 'lambda' function to each element of 'nums'.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Answer:**Difference Between 'map()', 'reduce()', and 'filter()' in Python:**  

1. **map(function, iterable)**:  
   - Applies a function to **each item** of an iterable.  
   - Returns an iterator with the transformed items.  

2. **filter(function, iterable)**:  
   - Applies a function to **each item** and keeps only those that return 'True'.  
   - Returns an iterator with the filtered items.  

3. **reduce(function, iterable)** (from 'functools' module):  
   - Applies a function **cumulatively** to the items of an iterable.  
   - Returns a **single value**.  
**Example:**  

from functools import reduce  

nums = [1, 2, 3, 4]

* map: squares each element  
print(list(map(lambda x: x**2, nums)))  # [1, 4, 9, 16]  

* filter: keeps only even numbers  
print(list(filter(lambda x: x % 2 == 0, nums)))  # [2, 4]  

* reduce: sums all elements  
print(reduce(lambda x, y: x + y, nums))  # 10

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

#check prime number

def is_prime(n):
    #Checks if a given number is prime or not.

    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Check prime numbers from 1 to 200

prime_numbers = [num for num in range(1, 201) if is_prime(num)]


print("Prime numbers from 1 to 200:")

print(prime_numbers)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

#Fibonacci sequence

class Fibonacci:
    #Iterator class to generate Fibonacci sequence up to a specified number of terms.

    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration

        if self.count == 0:
            result = 0
        elif self.count == 1:
            result = 1
        else:
            result = self.current + self.next
            self.current, self.next = self.next, result

        self.count += 1
        return result

# Example

num_terms = 30

fib = Fibonacci(num_terms)

print(f"Fibonacci sequence up to {num_terms} terms:")

print(list(fib))
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

# Power of Two Generator

def powers_of_two(exponent):
    #Generator that yields the powers of 2 up to the given exponent.

    for i in range(exponent + 1):
        yield 2 ** i

# Example

max_exponent = 10
print(f"Powers of 2 up to 2^{max_exponent}:")
for power in powers_of_two(max_exponent):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

#File line reader

def read_file_line_by_line(filename):
    #Generator function that reads a file line by line and yields each line as a string.

    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

# Create a demo text file

with open('demo.txt', 'w') as f:
    f.write("Line 1: Hello, world!\n")
    f.write("Line 2: Python is fun.\n")
    f.write("Line 3: Generators are efficient.\n")


filename = 'demo.txt'

try:
    for line in read_file_line_by_line(filename):
        print(line)

except FileNotFoundError:
    print(f"File '{filename}' not found.")
```

```python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

#sort a list of tuples

# List of tuples
items = [(1, 5), (3, 2), (4, 8), (2, 1)]

# Sorting using lambda function based on the second element
sorted_items = sorted(items, key=lambda x: x[1])

# Display the sorted list
print("Sorted list based on second element:")
print(sorted_items)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

#From Celsius to Fahrenheit

# List of temperatures in Celsius

temps_celsius = [0, 30, 40, 75]

# Using map() to convert to Fahrenheit

temps_fahrenheit = list(map(lambda c: (c * 9/5) + 32, temps_celsius))

# Display the converted temperatures

print("Temperatures in Fahrenheit:")
print(temps_fahrenheit)

```
